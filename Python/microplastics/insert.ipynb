### 方法1:  仅删除无数据行+ KNN填充缺失数据

```python
# 第一步：提取数值列和非数值列
non_numeric_cols = ['state', 'birth_year', 'birth_month', 'LBW_ratio', 'latitude', 'longitude']
numeric_cols = ['mp1', 'mp2', 'mp3', 'mp4', 'mp5']

# 第二步：删除缺失值大于等于 4 个的行（仅判断数值列）
df_filtered = df[df[numeric_cols].isnull().sum(axis=1) < 5]

print("\n删除缺失值大于等于4个的行后：")
print(df_filtered)
```

```python
# 第三步：用 KNN 方法填充数值列的缺失值
imputer = KNNImputer(n_neighbors=3)
numeric_filled = imputer.fit_transform(df_filtered[numeric_cols])

# 第四步：将填充后的数值列与非数值列合并
df_filled = pd.DataFrame(numeric_filled, columns=numeric_cols, index=df_filtered.index)
df_filled[non_numeric_cols] = df_filtered[non_numeric_cols]

# 重新排列列的顺序
df_result = df_filled[non_numeric_cols + numeric_cols]

print("\n最终结果（缺失值填充后）：")
print(df_result)
```

```python
# 检查是否还有缺失值
print("\n各列缺失值统计：")
print(df_result.isnull().sum())
```

```python
df_result.to_csv("method1.csv", index=False)
```

### 方法2:  仅删除无数据行+ 线性插值填充缺失数据
remark: 因为保留了仅有一个数据的行，所以仅有一个数据的行用线性插值后，mp1=mp2=mp3=mp4=mp5

```python
# 第一步：分离数值列和非数值列
non_numeric_cols = ['state', 'birth_year', 'birth_month', 'LBW_ratio', 'latitude', 'longitude']
numeric_cols = ['mp1', 'mp2', 'mp3', 'mp4', 'mp5']

# 第二步：删除缺失值大于等于 4 个的行（仅判断数值列）
df_filtered = df[df[numeric_cols].isnull().sum(axis=1) < 5]

print("\n删除缺失值大于等于 4 个后的数据：")
print(df_filtered)
```

```python

# 第三步：按行用插值法填充缺失值
def interpolate_row(row):
    return row.interpolate(method='linear', axis=0, limit_direction='both')

# 对数值列按行插值
df_filled_numeric = df_filtered[numeric_cols].apply(interpolate_row, axis=1)

# 第四步：合并数值列和非数值列
df_filled = pd.concat([df_filtered[non_numeric_cols], df_filled_numeric], axis=1)

print("\n最终结果（缺失值按行插值法填充后）：")
print(df_filled)
```

```python
# 检查是否还有缺失值
print("\n各列缺失值统计：")
print(df_filled.isnull().sum())
```

### 方法3: 将所有的缺失值都填充为-0.000001
* 优点：对于某些机器学习模型（如决策树、随机森林），填充 -0.000001 可以提供一种显式的 "缺失状态"， 决策树类模型（如 XGBoost、LightGBM）对数据分布不敏感，-0.000001 的离群值可以用作区分特征的一个标志
* 缺点：对基于距离或分布的模型（如 KNN、线性回归）产生不利影响。对于决策树类模型影响较小

```python
df_filled = df.fillna(-0.000001)
df_filled
```

### 方法4: 删除无数据行+ 中位数填充

```python
# 第一步：分离数值列和非数值列
non_numeric_cols = ['state', 'birth_year', 'birth_month', 'LBW_ratio', 'latitude', 'longitude']
numeric_cols = ['mp1', 'mp2', 'mp3', 'mp4', 'mp5']

# 第二步：删除缺失值大于等于 4 个的行（仅判断数值列）
df_filtered = df[df[numeric_cols].isnull().sum(axis=1) < 5]

print("\n删除缺失值大于等于 4 个后的数据：")
print(df_filtered)
```

```python
# 计算每行的中位数，并用中位数填充缺失值
def fill_with_row_median(row):
    row_median = row.median()  # 计算每行的中位数（忽略 NaN）
    return row.fillna(row_median)  # 用中位数填充缺失值

# 对数值列逐行应用填充
df_filled = df_filtered[numeric_cols].apply(fill_with_row_median, axis=1)

# 合并非数值列和填充后的数值列，保持原表不修改
df_result = pd.concat([df_filtered[non_numeric_cols], df_filled], axis=1)

print("填充后的数据：")
print(df_result)
```

```python
# 检查是否还有缺失值
print("\n各列缺失值统计：")
print(df_result.isnull().sum())
```

```python
df_result.to_csv("method4.csv", index=False)
```

### 方法5:删除无数据行+ 均值填充

### 多项式插值、样条插值至少都需要3个以上的有效数据，由于每行有效数据不够，无法达成有效插值，暂放弃

```python
import numpy as np
from scipy.interpolate import lagrange
```

```python
# 第一步：分离数值列和非数值列
non_numeric_cols = ['state', 'birth_year', 'birth_month', 'LBW_ratio', 'latitude', 'longitude']
numeric_cols = ['mp1', 'mp2', 'mp3', 'mp4', 'mp5']

# 第二步：删除缺失值大于等于 4 个的行（仅判断数值列）
df_filtered = df[df[numeric_cols].isnull().sum(axis=1) < 5]

print("\n删除缺失值大于等于 4 个后的数据：")
print(df_filtered)
```

```python
from scipy.interpolate import CubicSpline

# 第三步：按行用样条插值法填充缺失值
def spline_interpolation(row):
    y = row.values  # 获取当前行的值
    x = np.arange(len(y))  # 列索引
    non_nan_mask = ~np.isnan(y)  # 非缺失值的布尔掩码
    
    # 检查非缺失值数量，至少需要 2 个非缺失值
    if non_nan_mask.sum() > 2:
        cs = CubicSpline(x[non_nan_mask], y[non_nan_mask])  # 构建样条函数
        y_filled = np.where(np.isnan(y), cs(x), y)  # 用样条插值填充缺失值
        return y_filled
    return y  # 如果非缺失值不足，返回原数据

# 对数值列逐行应用样条插值
df_filled_numeric = df_filtered[numeric_cols].apply(spline_interpolation, axis=1, result_type="expand")

# 第四步：合并数值列和非数值列
df_filled = pd.concat([df_filtered[non_numeric_cols].reset_index(drop=True),
                       pd.DataFrame(df_filled_numeric, columns=numeric_cols).reset_index(drop=True)], axis=1)

print("\n最终结果（缺失值按行样条插值法填充后）：")
print(df_filled)
```
