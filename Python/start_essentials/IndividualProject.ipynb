### Part 2
This is a budget control program, the algorithm of the given code below is :

1.	Get the manager to enter the budget allocated to a project

2.	Accept all claims and less off the amount from the budget as balance.
 
3.	Control the budget until no more balance.


```python
# 2 Given Code
# Get the initial budget allocated to the project from the manager
budget = float(input("Enter the budget allocated to the project: "))

# Initialize balance with the initial budget
claim = 0

# Loop to accept claims until the budget is exhausted
while budget > claim:
    # Get the amount of the claim
    claim = float(input("Enter the amount of the claim: "))

    # Check if the claim exceeds the available balance
    if claim > budget:
        print("Claim exceeds available balance. Please enter a smaller claim.")

    # Deduct the claim amount from the balance
    budget -= claim

    print(f"Claim accepted. Remaining balance: {budget}")

print("Budget exhausted. No more claims can be accepted.")
```

```python
# Insert generative codes here
def budget_control():
    # Step 1: Get the manager to enter the budget allocated to a project
    budget = float(input("Enter the budget allocated to the project: "))
    balance = budget

    print(f"Initial budget: {budget:.2f}")

    # Step 2: Accept all claims and deduct the amount from the budget
    while balance > 0:
        claim = float(input("Enter a claim amount (or 0 to stop): "))
        
        # Stop if claim is 0
        if claim == 0:
            break
        
        if claim <= balance:
            balance -= claim
            print(f"Claim accepted. Remaining balance: {balance:.2f}")
        else:
            print("Claim exceeds remaining budget. Claim rejected.")

    # Step 3: Control the budget until no more balance
    if balance == 0:
        print("Budget exhausted.")
    else:
        print(f"Final balance: {balance:.2f}")

budget_control()
```

The code provided is a program prepared to restore a crashed database.
An example of how this program works is as follows:
If the stock balances of fruits as at last backed up were:
apple => 20

And the transaction log from the previous backed up till system crash is as follows:
apple, sold, 10

durian, stock-up, 90
durian, sold, 1
papay, sold, 2
durian, sold, 1

The given program will work out the current balances of the fruits from the previous balance and the transaction log. For example, apple will be 20 (balance) – 10 (sold)  + 50 (stock-up)  - 2(sold) = 58. 
In the transaction log, you may assume that there are only 2 activities, sold or stock-up. However, the fruits appear in the backed-up balance may not be exhaustive.


```python
# 4 Given Code
apple = 20
orange = 40
papaya = 60

log = '''apple, sold, 10
orange, stock-up, 20
apple, stock-up, 50
durian, stock-up, 90
durian, sold, 1
papaya, sold, 2
durian, sold, 1
apple, sold, 2
papaya, stock-up, 20'''

for transaction in log.split('\n'):
    [fruit, action, qty] = transaction.split()
    if fruit == 'apple':
        if action == 'sold':
            apple -= qty
        else:
            apple += qty
    if fruit == 'orange':
        if action == 'sold':
            orange -= qty
        else:
            orange += qty
    if fruit == 'papaya':
        if action == 'sold':
            papaya -= qty
        else:
            papaya += qty
    if fruit == 'durian':
        if action == 'sold':
            durian = qty
        else:
            durian = qty
print("New balance")
print("Apple", apple)
print("Orange", orange)
print("Papaya", papaya)
#print("Duiran", durian)
```

```python
# Insert your proposed code here
# version 1
apple = 20
orange = 40
papaya = 60
durian = 0

log = '''apple, sold, 10
orange, stock-up, 20
apple, stock-up, 50
durian, stock-up, 90
durian, sold, 1
papaya, sold, 2
durian, sold, 1
apple, sold, 2
papaya, stock-up, 20'''

for transaction in log.split('\n'):
    fruit = transaction.split(',')[0].strip()
    action = transaction.split(',')[1].strip()
    qty = int(transaction.split(',')[2].strip())
    if fruit == 'apple':
        if action == 'sold':
            apple -= qty
        else:
            apple += qty
    if fruit == 'orange':
        if action == 'sold':
            orange -= qty
        else:
            orange += qty
    if fruit == 'papaya':
        if action == 'sold':
            papaya -= qty
        else:
            papaya += qty
    if fruit == 'durian':
        if action == 'sold':
            durian -= qty
        else:
            durian += qty
print("New balance")
print("Apple", apple)
print("Orange", orange)
print("Papaya", papaya)
print("Duiran", durian)
```

```python
# Insert generative codes here
# Initialize the backed-up stock balances
stock_balances = {
    'apple': 20,
    'orange': 40,
    'papaya': 60
}

# Transaction log
transaction_log = '''
apple, sold, 10
orange, stock-up, 20
apple, stock-up, 50
durian, stock-up, 90
durian, sold, 1
papaya, sold, 2
durian, sold, 1
apple, sold, 2
papaya, stock-up, 20
'''

# Process each transaction in the log
for transaction in transaction_log.strip().split('\n'):
    fruit, activity, quantity = transaction.split(', ')
    quantity = int(quantity)

    if fruit not in stock_balances:
        # Initialize the fruit balance if not present in backed-up stock
        stock_balances[fruit] = 0

    if activity == 'sold':
        stock_balances[fruit] -= quantity
    elif activity == 'stock-up':
        stock_balances[fruit] += quantity

# Print the current balances
print("Current stock balances:")
for fruit, balance in stock_balances.items():
    print(f"{fruit}: {balance}")
```
